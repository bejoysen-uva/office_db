```python
def link_dicts_2(parent_df, child_df, parent_col, child_col):
    '''This function will take in a parent dict and child dataframe, where
    the child dataframe will be nested inside a parent dict. The aim
    of this function is to prepare dictionaries to be uploaded to
    nonrelational databases, such as mongodb.'''
    parent_copy = parent_df.copy(deep = True)
    new_col_list = []
    for i in range(0, len(parent_copy)):
        child_id = parent_copy[parent_col].iloc[i] #linking child and parent id's
        child_info = child_df[child_df[child_col] == child_id]
        child_info = child_info.to_dict('records')
#         parent_copy[parent_col].iloc[i] = child_info
        if (len(child_info)) < 1:
            new_col_list.append(child_info)
            print(child_info)
        else:
            new_col_list.append(child_info[0])
            print(child_info)
    return new_col_list

def link_dicts_3(parent_df, child_df, parent_col, child_col):
    '''This function will take in a parent dict and child dataframe, where
    the child dataframe will be nested inside a parent dict. The aim
    of this function is to prepare dictionaries to be uploaded to
    nonrelational databases, such as mongodb.'''
    parent_copy = parent_df.copy(deep = True)
    new_col_list = []
    for i in range(0, len(parent_copy)):
        child_id = parent_copy[parent_col].iloc[i] #linking child and parent id's
        child_info = child_df[child_df[child_col] == child_id]
        child_info = child_info.to_dict('records')
#         parent_copy[parent_col].iloc[i] = child_info
        new_col_list.append(child_info)
    return new_col_list
        
        
def link_dicts(parent_dict, child_df, parent_col, child_col):
    '''This function will take in a parent dict and child dataframe, where
    the child dataframe will be nested inside a parent dict. The aim
    of this function is to prepare dictionaries to be uploaded to
    nonrelational databases, such as mongodb.'''
    parent_copy = parent_dict.copy()
    for i in range(0, len(parent_dict)):
        child_id = parent_copy[i][parent_col] #linking child and parent id's
        child_info = child_df[child_df[child_col] == child_id]
        child_info = child_info.to_dict('records')
        parent_copy[i][parent_col] = child_info[0]
    return parent_copy
        
#         class_id = student_records_df[i]['Class'] # Getting the class id
#     class_info = class_df[class_df['Class'] == class_id][['Name', 'Prof']].iloc[0].to_dict() # Finding the corresponding class in the class dataframe and casting it to a dict
    
#     student_records_df[i]['Class'] = class_info # Updating the list above to replace class id with the dictionary created above
        
```
