### Setting things up

```python
import pymongo
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient('localhost', 27017)
```

```python
client
```

```python
db = client.example
```

```python
collection = db.students
```

### Inserting/deleting data

```python
student = {'name': 'Hayden', 'lname': 'French', 'year': 3, 'major': 'math', 'is_undergrad': True}
```

```python
collection.insert_one(student)
```

```python
collection.delete_one({'name':'Hayden'})
```

```python
students = [{'name': 'Hayden', 'lname': 'French', 'year': 3, 'major': 'math', 'is_undergrad': True},
           {'name': 'Silas', 'lname': 'Hayes', 'year': 3, 'major': 'stats', 'is_undergrad': True},
           {'name': 'Anya', 'lname': 'Pfeiffer', 'year': 5, 'major': 'education', 'is_undergrad': False}]
```

```python
collection.insert_many(students)
```

```python
collection.delete_many({'is_undergrad':True})
```

```python
collection.delete_many({})
```

### Querying data

```python
employees = pd.read_csv('https://raw.githubusercontent.com/hmf9kx/LDS2022/main/week2/employees.csv')
employees
```

```python
employees_dict = employees.to_dict('records')
employees_dict
```

```python
employee_collection = db.employees
```

```python
employee_collection.insert_many(employees_dict)
```

```python
query = {'first_name': 'Adam'}
employee_collection.find_one(query)
```

```python
query = {'team': 'A'}
employee_collection.find(query)
```

```python
query = {'team': 'A'}
result = employee_collection.find(query)
```

```python
for e in result:
    print(e)
```
